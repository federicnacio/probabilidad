tablas de frecuencias

```python
import random
Edades = []
for I in range(100):
 número_aleatorio = random.randint(18, 70)
 Edades.append(número_aleatorio)
print(Edades)
```

```python
import pandas as pd
```

```python
Df = pd.DataFrame(Edades, columns=['Edad'])
Df.head()
```

```python
tabla_frecuencias = Df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad','Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by='Edad')

print("Tablas de distribución de frecuencias:")
tabla_frecuencias
```
