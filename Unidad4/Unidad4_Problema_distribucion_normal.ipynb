Una revisión de los pesos de 500 estudiantes del itd

```python
import numpy as np
import matplotlib.pyplot as PLT
from scipy.stats import norm
```

```python

media=70
desv_std=4
num_datos=500
datos = np.random.normal (media, desv_std, num_datos)
datos[0:50]
```

```python
x=np.linspace(media - 3 * desv_std, media + 3 * desv_std, num_datos)

pdf=1 / (desv_std * np.sqrt(2 * np.pi)) * np.exp( - (x - media) ** 2 / (2 * desv_std ** 2))
```

```python
print("los primeros 50 valores de x", x[0:50], " , y los primeros 50 pdf", pdf[0:50])
```

```python
PLT.figure(figsize=(6, 4))
PLT.hist(datos, bins=15, alpha=0.4, color='b')
PLT.title('distribución normal')
PLT.xlabel('peso en Kg de los estudiantes')
PLT.grid(True)
PLT.show()
```

```python
PLT.figure(figsize=(6, 4))
PLT.plot(x, pdf, label='nornal: (media:{}; ds:{})'.format(np.round(media, 2), np.round(desv_std, 2)))
PLT.title('funcion de distribucion de probabilidad (pdf) - distribucion normal')
PLT.xlabel('pesos en kg. de los estudiantes')
PLT.ylabel('densidad de probabilid')
PLT.legend()
PLT.grid(True)
PLT.show()
```

```python
PLT.figure(figsize=(6, 4))
PLT.hist(datos, bins=15, alpha=0.4, density=True, color='b')
PLT.plot(x, pdf, 'k', linewidth=2, )
PLT.title('distribucion normal')
PLT.xlabel('peso en kg. de los estudiantes')
PLT.grid(True)
PLT.show()
```

Problema 1 Ditribucion de pesos

- probabibilidad de que los estudiantes tengan un peso mayor a 68 y menor a 72
- numero de estudiantes dentro de ese rango de peso


```python
media = 70
desv_std = 4
total_estudiantes = 500
peso_limite = 79
```

```python
prob_68_72 = norm.cdf(72, loc=media, scale=desv_std) - norm.cdf(68, loc=media, scale=desv_std)
num_estudiantes_68_72 = int(prob_68_72 * total_estudiantes)

print(f"La probabilidad de que un estudiante pese más de 68 kg y menos de 72 kg es: {prob_68_72:.4f}")
print(f"Se estima que {num_estudiantes_68_72} estudiantes pesan más de 68 kg y menos de 72 kg.")
```
