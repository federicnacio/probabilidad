```python
from statistics import *
import random
```

```python
fabrica1=[]
for i in range(100):
 numero_aleatorio=round(random.uniform(50,100),2)
 fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2=[]
for i in range(100):
 numero_aleatorio=round(random.uniform(50,100),2)
 fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("produccion diaria fabrica1")
```

```python
media1=mean(fabrica1)
mediana1=median(fabrica1)
moda1=mode(fabrica1)
print("media=", media1," ","mediana=", mediana1," ", "moda=", moda1)
```

```python
plt.hist(fabrica2)
plt.title("produccion diaria fabrica2")
```

```python
media2=mean(fabrica2)
mediana2=median(fabrica2)
moda2=mode(fabrica2)
print("media=", media2," ","mediana=", mediana2," ", "moda=", moda2)
```

```python
fig,(ax1,ax2)=plt.subplots(nrows=1,ncols=2,figsize=(10,5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```

```python
plt.hist(fabrica1,bins=10,alpha=0.2,color='b',edgcolor="black")
plt.axvline(media1,color='b',linestyle='dashed',linewidth=2,label='media')
plt.axvline(mediana1,color='r',linestyle='dashed',linewidth=2,label='mediana')
plt.axvline(moda1,color='y',linestyle='dashed',linewidth=2,label='moda')
plt.xlabel('valor')
plt.ylabel('frecuencia')
plt.title("produccin diaria de la fabrica 1")
plt.legend()
plt.show()
```
